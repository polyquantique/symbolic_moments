# LoopMontrealer Investigation

The loopMontrealer is the Montrealer allowing for displacement, but without repetition.

```python
import symbolic_cumulants as gbs
import numpy as np
from sympy import symbols, Matrix
```

```python
def double_factorial(n):
     if n == 0:
         return 1
     else:
         return n * double_factorial(n-2)

for n in range(1,5):
    modes = {i:1 for i in range(1,n+1)} #No repetition
    zeta = symbols("alpha1:%d" % (n+1)) #With displacement
    zeta = np.array(zeta)
    zeta = np.concatenate((zeta, zeta.conj()))
    A = gbs.block_A(n, 1)

    loopmtl = gbs.photon_number_cumulant(A, zeta, modes)
    terms = len(loopmtl.args)
    facto = double_factorial(2*n-2)
    print(f'Order : {n}, (2n-2)!! : {facto}, ', end=' ')
    print(f'loopMtl : {terms}, factor : {terms/facto}')
    display(loopmtl)

```
